#Оптимизация. Задача упаковки кругов#

Задача заключается в том, чтобы упаковать в единичный круг заданного числа $n$ равных друг другу меньших кругов радиуса $r$, так, чтобы радиус $r$ был максимальным. При этом нужно учитывать все условия и ограничения.












Подключим все необходимые нам библиотеки и создадим массив для найденных решений.


```python
import numpy as np
from scipy.optimize import minimize
import random as rd
import matplotlib.pyplot as plt
%matplotlib inline

mas = np.array([]) #массив для найденных решений
```


Напишем функцию, которая будет учитывать все ограничения и условия.


```python
def cons(x):
    mas1 = np.array([])
    for j in range(1, 2*n, 2):
        mas1 = np.append(mas1, (1-x[0])**2 - x[j]**2 - x[j+1]**2)
    for i in range(1, 2*n, 2):
        for j in range(i+2, 2*n+1, 2):
            mas1 = np.append(mas1, (x[i] - x[j])**2 + (x[i+1] - x[j+1])**2 - 4)
    mas1 = np.append(mas1, x[0] - 1)
    mas1 = np.append(mas1, x[0])
    return mas1
```


Напишем функцию, принимающую на вход $n$ и находящую оптимальную упаковку. Также функция будет отрисовывать найденную упаковку.


```python
def optimal(n):
    mas2 = np.array([rd.uniform(-n,n) for i in range(2*n+1)])
    mas2[0] = n
    m = minimize(lambda x: x[0], mas2, constraints=({'type':'ineq','fun': cons}))
    x = m.x / m.x[0]
    r = x[0] / m.x[0]
    ax=plt.gca()
    ax.add_patch(plt.Circle((0,0), 1, fill=False))
    plt.axis('scaled')
    for i in range(1, 2*n+1, 2):
        ax=plt.gca()
        ax.add_patch(plt.Circle((x[i],x[i+1]), r, fill='blue'))
        plt.axis('scaled')
    plt.show()
    global mas
    mas = np.append(mas, r)
    print("При n =",n, "радиус равен:", r)
    return 
```
